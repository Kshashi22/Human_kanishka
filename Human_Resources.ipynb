```python
MartialStatus = data['MaritalStatus'].value_counts(normalize=True) # Changed 'MartialStatus' to 'MaritalStatus'
print(MartialStatus)

plt.figure(figsize=(5,4))
plt.pie(MartialStatus.values,labels=MartialStatus.index,autopct='%1.1f%%',startangle=90)
plt.title('Marital Status of the employees') # Changed title to reflect the correct column name
plt.show()
```

```python
print(data.columns)
NumCompaniesWorked = data['NumCompaniesWorked'].value_counts(normalize=True)
print(NumCompaniesWorked)

plt.figure(figsize=(5,4))
plt.pie(NumCompaniesWorked.values,labels=NumCompaniesWorked.index,autopct='%1.1f%%',startangle=90)
plt.title('NumCompaniesWorked')
plt.show()
```

```python
Education = data['Education'].value_counts(normalize=True)
print(Education)

plt.figure(figsize=(5,4))
plt.pie(Education.values,labels=Education.index,autopct='%1.1f%%',startangle=90)
plt.title('Education of the employees')
plt.show()
```

```python
print(data.columns)
Attrition= data['Attrition'].value_counts(normalize=True)
print(Attrition)

plt.figure(figsize=(5,4))
plt.pie(Attrition.values,labels=Attrition.index,autopct='%1.1f%%',startangle=90)
plt.title('Attrition')
plt.show()
```

```python
print(data.columns)
BusinessTravel= data['BusinessTravel'].value_counts(normalize=True)
print(BusinessTravel)

plt.figure(figsize=(5,4))
plt.pie(BusinessTravel.values,labels=BusinessTravel.index,autopct='%1.1f%%',startangle=90)
plt.title('BusinessTravel')
plt.show()
```

```python
print(data.columns)
Overtime = data['OverTime'].value_counts(normalize=True)
print(Overtime)

plt.figure(figsize=(5,4))
plt.pie(Overtime.values,labels=Overtime.index,autopct='%1.1f%%',startangle=90)
plt.title('Overtime done by employees')
plt.show()
```

```python
plt.figure(figsize=(27,8))
sns.countplot(x='JobRole',hue='JobSatisfaction',data=data,palette='rocket')
plt.title('Job Role and Job Satisfaction')
plt.xlabel('Job Role')
plt.ylabel('Job Satisfaction')
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x='Department',hue='PerformanceRating',data=data,palette='rocket')
plt.title('Department and PerformanceRating')
plt.xlabel('Department')
plt.ylabel('PerformanceRating')
```
